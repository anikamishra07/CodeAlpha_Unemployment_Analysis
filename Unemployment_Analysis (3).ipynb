## Data Cleaning

```python
df = df.dropna()
```

```python
df.describe()
```

```python
avg_unemployment = df["Estimated Unemployment Rate (%)"].mean()

print("Average Unemployment Rate:", avg_unemployment)
```

```python
state_unemployment = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean()

state_unemployment
```

```python
state_unemployment.sort_values(ascending=False).head()
```

```python
import matplotlib.pyplot as plt
```

## Data Visualization

Visualized unemployment trends using graphs for better understanding.

```python
state_unemployment.sort_values().plot(kind="bar", figsize=(12,5))

plt.title("State wise Average Unemployment Rate")
plt.xlabel("State")
plt.ylabel("Unemployment Rate (%)")

plt.show()
```

```python
area = df.groupby("Area")["Estimated Unemployment Rate (%)"].mean()

area
```

```python
top_states = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean().sort_values(ascending=False).head(10)

top_states.plot(kind="bar")

plt.title("Top 10 States with Highest Unemployment Rate")
plt.xlabel("States")
plt.ylabel("Unemployment Rate (%)")

plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
date_trend = df.groupby("Date")["Estimated Unemployment Rate (%)"].mean()

date_trend.plot(kind="line")

plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")

plt.show()
```

```python
employment = df.groupby("Region")["Estimated Employed"].mean().sort_values(ascending=False).head(10)

employment.plot(kind="bar")

plt.title("Top 10 States by Employment")
plt.show()
```

```python
area.plot(kind="bar")

plt.title("Rural vs Urban Unemployment")

plt.show()
```
